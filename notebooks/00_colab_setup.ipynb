```python
import os
import sys
from pathlib import Path

REPO_URL = "https://github.com/zkc768/hf_stock_clf.git"
REPO_DIR = Path("/content/hf_stock_clf")

# Clone or pull
if REPO_DIR.exists():
    get_ipython().system('cd {REPO_DIR} && git pull --ff-only')
else:
    get_ipython().system('git clone {REPO_URL} {REPO_DIR}')

if str(REPO_DIR) not in sys.path:
    sys.path.insert(0, str(REPO_DIR))

print(f"Repo ready: {REPO_DIR}")
```
